```python
import requests
from lxml.html import fromstring
import requests
import time
import json
from json import JSONDecodeError
from torpy.http.requests import TorRequests
import random

class vote_cheater_ONE:
    
    def __init__(self,survey_num,candidate_num):
        
        self.survey_num = survey_num
        self.candidate_num = candidate_num
        
    def vote(self):
        
        vote_cheater_ONE.vote_for_candidate(self.survey_num,self.candidate_num)
        
        
    def get_candidate_votes_from_text(txt):
        try:
            res = json.loads(txt)['Answers'][0]['Votes']
            print(res)
            return res
        except JSONDecodeError:
            print(f'could not parse json')

    def override_params(params,candidate_num):
        return {
            'vId': params['vId'],
            'a': 'vote',
            'aId': str(candidate_num),
            '_': vote_cheater_ONE.get_now()
        }

    def get_now():
        return ''.join(str(time.time()).split('.'))[:13]


    def vote_for_candidate(survey_num,candidate_num):

        get_poll = 'https://counter.one.co.il/Cat/General/AjaxHandler.ashx'
        params = {
            'vId': str(survey_num),
            '_': vote_cheater_ONE.get_now(),
            'a': 'get-vote'
        }
        with TorRequests() as tor_requests:
            with tor_requests.get_session() as sess:
                sess.cookies.clear()
                res = sess.get(url=get_poll, params=params)
                time.sleep(3.0)

                vote_params = vote_cheater_ONE.override_params(params,candidate_num)
                vote_res = sess.get(url=get_poll, params=vote_params)
                time.sleep(3.0)

                print(f'number of votes before {vote_cheater_ONE.get_candidate_votes_from_text(res.text)} after {vote_cheater_ONE.get_candidate_votes_from_text(vote_res.text)}')
```
